```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary
    
    def display_salary(self):
        return f"salary is : {self.salary}"
    
    def increase_salary(self, new_salary):
        self.salary = new_salary
        return f"{self.name}'s salary is now {self.salary}"
    
e1 = Employee("Charvak", 73500)
print(e1.display_salary())
e1.increase_salary(88300)
```

```python
class BankAccount:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, amount):
        if amount >0:
            self.balance += amount
            return f"Deposit Rupee {amount}. New Balance is {self.balance}"
        
    def withdraw(self, amount):
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            return f"withdraw Rupee {amount}. currant Balance is {self.balance}"
        return f"Insufficient funds of Invalid amount"
    
    def show_balance(self):
        return f"currant Balance is {self.balance}"
    
    def display_info(self):
        return f"""Account Holder name : {self.account_holder}
Balance is {self.balance}"""
    
ah1 = BankAccount("Charvak", 1000000)
print(ah1.display_info())
print(ah1.show_balance())
ah1.deposit(50000)
ah1.withdraw(80000)
```

```python
import math

class circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        circle_area = math.pi * self.radius **2
        return f"area of circle is {circle_area}"
    
    def circumference(self):
        circle_circum = 2 * math.pi * self.radius
        return f"Circumference of Circle is {circle_circum}"
    
R1 = circle(3.4)
R1.area()
R1.circumference()
print(R1.area())
print(R1.circumference())
```

```python
class Laptop:
    def __init__(self, brand, price):
        self.brand = brand
        self.price = price

    def apply_discount(self, percent):
        Discount = self.price * (1 - percent / 100)
        return f" Discount on laptop is : {Discount}"
    
    def show_details(self):
        return f"Laptop Brand: {self.brand}, Price = {self.brand}"
    
l1 = Laptop("Lenovo LOQ", 120000)
print(l1.show_details())
print(l1.apply_discount(30))
```

```python
class movie:
    def __init__(self, title, rating):
        self.title = title
        self.rating = rating

    def is_hit(self):
        if self.rating >= 8:
            return f"{self.title} Movie is Hit with rating of {self.rating}"
        return f"{self.title} Movie is Aveage with rating of {self.rating} "
    
m1 = movie("interstellar", 9.3)
m2 = movie("Bhooth Bangla", 6)

print(m1.is_hit())
print(m2.is_hit())
```

```python
class library:
    def __init__(self):
        self.books =[]

    def add_book(self, book_name):
        self.books.append(book_name)
    
    def show_books(self):
        return f"Listed Books are : {self.books}"
    
b1 = library()
b1.add_book("book1")
b1.add_book("book2")
print(b1.show_books())
```
