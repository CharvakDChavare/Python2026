Modify Properties

```python
class person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

p1 = person("charvak", 23)
print(p1.age)
p1.name="Manav"
print(p1.name)
```

Delete properties

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

p1 = Person("charvak", 30)

del p1.age

print(p1.name) # This works
# print(p1.age) # This would cause an error
```

Class Properties vs Object Properties

```python
class Person:
  species = "Human" # Class property

  def __init__(self, name):
    self.name = name # Instance property

p1 = Person("charvak")
p2 = Person("manav")

print(p1.name)
print(p2.name)
print(p1.species)
print(p2.species)
```

Modifying Class Properties

```python
class Person:
  lastname = ""

  def __init__(self, name):
    self.name = name

p1 = Person("charvak")
p2 = Person("Manav")

Person.lastname = "chavare"

print(p1.lastname)
print(p2.lastname)
```

Add New Properties

```python
class Person:
  def __init__(self, name):
    self.name = name

p1 = Person("charvak")

p1.age = 23
p1.city = "Mohol" \

print(p1.name)
print(p1.age)
print(p1.city)
```
