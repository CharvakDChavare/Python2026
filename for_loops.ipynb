For LOOP

```python
fruits = ["apple", "Banana", "Cherry"]
for fruit in fruits:
    print(fruit)
```

```python
for x in "banana":
    print(x)
```

The break Statement

```python
x = ["apple", "Banana", "Cherry"]
for x in fruits:
     print(x)
     if x == "Banana":
        break
    
```

The continue Statement

```python
x = ["apple", "Banana", "Cherry"]
for x in fruits:
     
     if x == "Banana":
        continue
     print(x)
```

The range() Function

```python
for x in range(6):
    print(x)
```

```python
for x in range(2, 6):
    print(x)
```

```python
for x in range(2, 20, 3):
    print(x)
```

Else in For Loop

```python
for x in range(6):
    print(x)

else:
    print("fineshed!")
```

Nested Loops

```python
adj = ["red", "big", "testy"] 
fruit = ["apple", "banana", "cherry"]

for x in  adj:
    for y in fruit:
        print(x, y)
```

```python
fruits = ["apple", "banana", "cherry"]
for x in fruits:
    if x == "banana":
        print(x)
        break
```

```python
# print numbers 1 to 100 using for loop

for x in range(1, 100):
    if x % 2 ==0:
        print(x)
```
