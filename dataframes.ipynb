GroupBy

```python
df.groupby('emp')['salary'].mean()
```

```python
g = df.groupby('city') # Common aggregations
print(g['salary'].sum(),"\n")
print(g['salary'].mean(),"\n")
print(g['salary'].max(),"\n")
print(g['salary'].count(),"\n")
print(g['salary'].std(),"\n")
```

```python
#Multiple aggregations
df.groupby('city').agg(
    avg_sal=('salary', 'mean'),
    headcount = ('emp', 'count'),
    max_sal = ('salary', 'max')
)
```

```python
df.groupby('city')['salary'].mean() # aggergate

df['city_avg'] = df.groupby('city')['salary'].transform('mean') # # transform → keeps original shape (row-aligned)
df['pct_of_ avg'] = df['salary'] / df['city_avg']
df
```

Merging

```python
employees = pd.DataFrame({'emp_id':[1,2,3], 'name':['alice','charlie','david']})
salaries  = pd.DataFrame({'emp_id':[1,2,3],  'pay': [70000, 75000, 80000]})
```

```python
pd.merge(employees, salaries, on='emp_id') # inner (default)
```

```python
pd.merge(employees, salaries , on='emp_id', how='left') # left
```

```python
pd.merge(employees, salaries , on='emp_id', how='outer') # outer
```

```python
pd.merge(employees, salaries , on='emp_id', how='right') # right
```

```python
pd.merge(employees, salaries ,  how='cross') # cross
```

```python
pd.concat([employees, salaries]) # stack rows axis = 0
```

```python
pd.concat([employees, salaries], axis=1) # stack rows axis = 1
```

```python
pd.concat([employees, salaries], ignore_index=True) # reset index after stacking
```

```python
pd.merge(employees , salaries, left_on='emp_id', right_on= 'pay')
```
