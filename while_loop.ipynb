Python While Loops

```python
i = 1
while i < 6:
    print(i)
    i += 1
```

The break Statement

```python
i = 1
while i < 6:
    print(i)
    if i == 3:
        break
    i += 1
```

The continue Statement

```python
i = 0
while i < 6 :
    i +=1
    if i ==3:
        continue
    print(i)
```

The else Statement

```python
i = 1
while i < 6:
    print(i)
    i +=1
else:
    print("i is no longer less than 6")
```

Infinite loop with break

```python
count = 0
while True:
    print(count)
    count += 1
    if count >= 5:
        break
```

Input loop with break

```python
while True:
    num = int(input("Enter the number:"))
    print("double of", num, "is", 2 *num)
    break
```
