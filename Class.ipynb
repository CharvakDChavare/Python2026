```python
class Persson:
    def __init__(self, name, age, city):

        self.name = name
        self.age =age
        self.city = city
        
    def introduce(self):
        return f"Hi, I'm {self.name} from {self.city}, and I'm {self.age} Years Old."
    
    def age_up(self, years=1):
        self.age += years
        return f"{self.name} is now {self.age} Years old."
    
p1 = Persson("Alice", 25, "Nashik")
p2 = Persson("Bob", 30, "Pune")

print(p1.name)
print(p1.introduce())
print(p1.age_up())
```

```python
class Student:
    def __init__(self, name, student_id, major, gpa=0.0):
        self.name = name
        self.student_id = student_id
        self.major = major
        self.gpa =gpa
        self.courses = []


    def add_course(self, course_name):
        self.courses.append(course_name)
        return f"{course_name} added to {self.name}'s schedule" 
    
    def update_gpa(self, new_gpa):
        self.gpa = new_gpa
        return f" {self.name} 's GPA is now {self.gpa}"
    
    def display_info(self):
        return f"""Student: {self.name} 
ID: {self.student_id}
Major: {self.major}
GPA:{self.gpa}
Courses: {self.courses}"""
    
s1 = Student("Charvak", "CS20098765", "Computer Science", 4.0)

print(s1.add_course("Data Engineer"))
print(s1.add_course("Data PipeLine"))
print(s1.update_gpa(4.5))
print(s1.display_info())
```

```python
class BankAccount:
    def __init__(self, owner, account_number, balance=0.0):
        self.owner= owner
        self.account_number = account_number
        self.balance = balance
    
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            return f"Deposited Rupe{amount}. New Balance: {self.balance}"
        return "Invalid amount"
    
    def withdraw(self, amount):
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            return f"Withdrawn Rupe{amount}. New balance: {self.balance}"
        return "Insufficient funds of Invalid amount"
    
    def get_balance(self):
        return f"Current balance: {self.balance}"
    
    def display_info(self):
        return f""" Owner: {self.owner}
account_number: {self.account_number}
"""
    

account = BankAccount("Alice", "1456786", 10000.0)
print(account.deposit(500))
print(account.withdraw(200))
print(account.withdraw(500))
print(account.get_balance())
print(account.display_info())
```

```python
class Product:
    def __init__(self, name, price, quantity, category):
        self.name = name
        self.price = price
        self.quantity = quantity
        self.category = category

    def update_price(self, new_price):
        self.price = new_price
        return f"{self.name}'s price is now {self.price}"
    
    def total_value(self):
        return self.price * self.quantity



    
    def get_info(self):
        return f""" Name: {self.name}
Price: {self.price}
Quantity: {self.quantity}
Category: {self.category}
Total Value:{self.total_value()}"""

p1 = Product("smart_Phone", 95000, 4, "Electronics")
print(p1.get_info())
print(p1.update_price(135000))
p2 = Product("Laptop", 88000, 6, "Electronics")
print("\n",p2.get_info())
print(p2.update_price(90000))
```

```python
class Animal:
    def __init__(self, name, age):
        self.name = name
        self.age = age

class Cat(Animal):
    def __init__(self, name, age, color):
        super().__init__(name, age)
        self.color = color
    
    def info(self):
        return f"""Name: {self.name}
Age: {self.age}
Color: {self.color}"""

cat1 = Cat("swity", 15, "white")
print(cat1.info())
```
